```python
import numpy as np
import pandas as pd

sales = pd.read_excel("Electronic sales.xlsx")
sales.info()
sales.head()
```

```python
sales.Sales.mean()
sales.Sales.median()
```

```python
sales["Order Date"].max()
```

```python
# Efficient summaries : the .agg() function
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)
```

```python
sales["Shipping Cost"].quantile(0.25)
```

```python
sales["Shipping Cost"].agg(max)
sales["Shipping Cost"].max()
```

```python
# the sales["Shipping Cost"] column is passed to iqr function as argument
print(sales["Shipping Cost"].agg(iqr))
```

```python
sales["Shipping Cost"].agg([np.median, np.mean, np.max, np.std])
```

```python
sales[["Shipping Cost", "Profit"]].agg([np.median, np.mean, np.max, np.std])
```

```python
# drop duplicates
sales.drop_duplicates(subset="Discount")
```

```python
# drop rows that have the same value for "Discount" and "Order Priority" 
sales.drop_duplicates(subset=["Discount", "Order Priority"])
```
