```python
import pandas as pd

homelessness = pd.read_csv("2013 US State Statistics.csv")

homelessness.info()
homelessness.head()
homelessness.shape
homelessness.describe()
```

```python
homelessness.values  # A two-dimensional NumPy array of values.
homelessness.columns
homelessness.index
```

# Sorting rows
homelessness.sort_values("population")

```python
homelessness.sort_values("Population", ascending=False).head()
```

```python
homelessness.sort_values("State", ascending=False).head()[["State", "Population"]]
```

```python
# subsetting cols
homelessness["State"]
homelessness.State
homelessness[["State", "Population"]].head()
```

```python
# Subsetting rows
# the idea is that homelessness["Population"] > 5742713 return a boolean column value(True | False), and by defaulte
# homelessness[bool_val] display rows that just have True value

homelessness[homelessness["Population"] > 5742713]
homelessness[homelessness.State == "Arizona"]
homelessness[(homelessness.Population > 5742713) & (homelessness.Population < 8742713)].head()
```

```python
# subsetting rows by categories
canu = ["California", "Arizona", "Nevada", "Utah"]
homelessness[homelessness.State.isin(canu)]
```
