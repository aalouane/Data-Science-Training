```python
import pandas as pd

temperatures = pd.read_csv("temperatures_1.csv")
print(temperatures.columns)
print(temperatures.index)
temperatures.head()
temperatures.info()
```

```python
temperatures = temperatures.rename(columns={"dt":"date"})
cols = ["Country","City", "date", "AverageTemperature", "AverageTemperatureUncertainty", "Latitude", "Longitude"]
temperatures = temperatures[cols]
temperatures.head()
```

```python
# set index column country
temperatures.set_index("Country")
```

```python
# reset index country
# with reset_index(), new column "index" will be add
temperatures.reset_index()

temperatures.reset_index(drop=True)
```

```python
# set_index columns City for subsetting
temperatures_ind = temperatures.set_index("City")
temperatures_ind.head()
```

```python
# subsetting 
cities = ["Ahmadabad", "Aleppo"]

# the first method
print(temperatures[temperatures["City"].isin(cities)])
## 9223   Syria     Aleppo  2013-09-01                 NaN
```

```python
# the second method, witch is more readable
temperatures_ind.loc[cities]
```

```python
# just the AverageTemperature > 24
# first method
print(temperatures[temperatures["AverageTemperature"] > 24])
```

```python
temperatures_ind = temperatures.set_index("AverageTemperature")
rows = [18,22]
temperatures_ind.loc[[18,22]] # or temperatures_ind.loc[values] : .loc[row, col] ==> .loc[[r1,r2] , [col1, col2]] 
```

```python
# subsetting rows and columns
cols = ["Country", "date"]
temperatures_ind.loc[rows , cols].info()
```

```python
print(temperatures[(temperatures["AverageTemperature"] >= 18 ) & ( temperatures["AverageTemperature"] <= 22) ])
```

```python
"""
Summarize between the two method:
rows = [18,22] | rows = [18,22,25,27]
with this temperatures_ind.loc[rows], we get the all rows that have the exact value in rows, 18 or/and 22 and not 18.102 or 22.141
==> 19 entries

to get range value : rows = [18:22], we must to sort the dataFrame : 
temperatures_ind = temperatures.set_index("AverageTemperature").sort_index()
temperatures_ind.loc[18:22]

so with this : temperatures_ind = temperatures.set_index("AverageTemperature").sort_index()
we can get exact value([18,22,35...]) or a range of value([18:25]) ==> the last index is included

with this 
print(temperatures[(temperatures["AverageTemperature"] > 18 ) & ( temperatures["AverageTemperature"] < 22) ])
here we define a condition, and we get the range value between 18 and 22 : 18.124 ..... 22.758
==> 35526 entries

Subsetting with set_index/values we get the exact value, on other hand, subsetting with index or condition we get range value
"""
```
